* Descriptive statistics
* Inferential statistics

```python
# Applied descriptive statistics to a specific column (i.e. max() maximum value)
forbes_df['Profits (Billions)'].max()
```

```python
# Filtered information in dataframe with one clause, predicate or condition
forbes_df[forbes_df['Profits (Billions)'] == forbes_df['Profits (Billions)'].max()]
```

```python
# Another way to get it
# idxmax() function returns the record index with the maximum value to Profits (Billions) = 3
forbes_df.loc[forbes_df['Profits (Billions)'].idxmax()]
```

```python
# Serie to Dataframe
forbes_df.loc[forbes_df['Profits (Billions)'].idxmax()].to_frame()
```

```python
# Transpose of Dataframe, with attribute T
forbes_df.loc[forbes_df['Profits (Billions)'].idxmax()].to_frame().T
```

```python
# Descriptive statistic for entire dataframe, operation is performed by column (i.e. mean() for all numeric columns)
forbes_df.mean(numeric_only=True)
```

```python
# Standard Deviation. The result indicates that, on average, the market value deviates from the mean by approximately 117 billions.
forbes_df['Market Value (Billions)'].std()
```

```python

# 75% of companies have profits (earnings) equals to or less than 2.27 billions
forbes_df['Profits (Billions)'].quantile(.75)
```

```python
# describe() function returns a summary of descriptive statistics for all numeric columns in dataframe.
forbes_df.describe().T
```

```python
# Descriptive statistics for all categorical/object columns in dataframe
forbes_df.describe(include='object').T
```

```python
# groupby() function makes partitions o groups, in every group all rows have the same value by the specific column
# i.e. get mean by industry for all numeric columns in dataframe
forbes_df.groupby('Industry').mean(numeric_only=True)
```

```python
# For a specific column, let's filter first, then perform groupby()
forbes_df[['Assets (Billions)', 'Industry']].groupby('Industry').mean()
```

```python
# For the previous command, dataframe values can be sorted in order ascending or descending
forbes_df[['Assets (Billions)', 'Industry']].groupby('Industry').mean().sort_values('Assets (Billions)', ascending=False)
```

```python
# It could have several levels of grouping
forbes_df[['Assets (Billions)','Industry','Country']].groupby(['Industry','Country']).mean()
```

```python
# plot() function has the parameter kind to specify the plot
# Every column represents a class or bin (which includes a range of values)
forbes_df['Revenue (Billions)'].plot(kind='hist')
```

```python
# Another way to plot the previous one
forbes_df['Revenue (Billions)'].plot.hist()
```

```python
# plot() function draws 10 bins by default
# To change the number of bins, add bins parameter
forbes_df['Revenue (Billions)'].plot(kind='hist', bins=20)
```

```python
# Also it can enumerate the limits for all bins, they do not require to be equidistant
forbes_df['Revenue (Billions)'].plot(kind='hist', bins=[0,25,50,100,200])
```

```python
# If limits require to be equidistant, let's use np.arange(start, end, increment) numpy function
import numpy as np
forbes_df['Revenue (Billions)'].plot.hist(bins=np.arange(0,200,25))
```

```python
# hist() function gets a histogram for each numeric column
# To adjust the shape size, add the parameter figsize
forbes_df.hist(figsize=(12,10))
```
